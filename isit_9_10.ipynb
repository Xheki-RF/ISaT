#Izom function 3D visualisation#

```python
import numpy as np
import math as m
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def izom_func(a):
  x, y = a
  return -1 * m.cos(x) * m.cos(y) * m.exp(-1 * ((x - m.pi) ** 2 + (y - m.pi) ** 2))
```

```python
def izom_1_var(x, y):
  return izom_func([x, y])
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

Z = np.zeros((len(x), len(y)))

for i in range(len(x)):
    for j in range(len(y)):
        Z[i, j] = izom_1_var(X[i, j], Y[i, j])

ax.plot_surface(X, Y, Z, cmap="jet")
plt.show()
```

```python
plt.contour(X, Y, Z)
plt.show()
```

```python
plt.scatter(X, Y, Z)
plt.show()
```

#Nelder-Mead optimization algorithm#

```python
from scipy import optimize
from geneticalgorithm import geneticalgorithm as ga
import pandas as pd
from geneticalgorithm import geneticalgorithm as ga
import pandas as pd
import time
```

```python
res = []
guess = [3, 3]
time_end_nm = []

for i in range(100):
  time_start_nm = time.time()

  root = optimize.minimize(izom_func, guess, method="Nelder-Mead", bounds=[(-3.5, 3.5), (-3.5, 3.5)])
  res.append(root.fun)

  time_end = time.time() - time_start_nm
  time_end_nm.append(time_end)
```

```python
print("Algorithm running time:", sum(time_end_nm))
```

```python
res
```

Calculating mathemetical waiting and standard deviation for time

```python
mean_nm = np.mean(time_end_nm)
std_nm = np.var(time_end_nm)
mean_nm, std_nm
```

Calculating mathemetical waiting and standard deviation for function

```python
mean_nm_f = np.mean(res)
std_nm_f = np.var(res)
mean_nm_f, std_nm_f
```

#Differential evolution algorithm#

```python
res_de = []
guess_de = [3, 3]

time_end_de = []

for i in range(100):
  time_start_de = time.time()

  root = optimize.differential_evolution(izom_func, bounds=[(-3.5, 3.5), (-3.5, 3.5)], init="random", strategy="best1exp")
  res_de.append(root.fun)

  time_end = time.time() - time_start_de
  time_end_de.append(time_end)
```

```python
print("Algorithm running time:", sum(time_end_de))
```

```python
res_de
```

#Calculating mathematical expectation and standard deviation#

For time

```python
mean_de = np.mean(time_end_de)
std_de = np.var(time_end_de)
```

```python
mean_de, std_de
```

For function

```python
mean_de_f = np.mean(res_de)
std_de_f = np.var(res_de)
```

#3D visualisation of correlation between function minimum and population size#

```python
res_ps_de = []
lst= []

for i in range(15):
  lst.append(i + 1)
  root = optimize.differential_evolution(izom_func, bounds=[(-3.5, 3.5), (-3.5, 3.5)], popsize=i+1, init="random", strategy="best1exp")
  res_ps_de.append(root.fun)
```

```python
import matplotlib.pyplot as plt

plt.plot(lst, res_ps_de, "c")
plt.grid()
plt.ylabel("Discovered minimum")
plt.xlabel("Population size")
plt.show()
```

#Uploading the dataset#


```python
from sklearn.datasets import load_digits

data = load_digits()
n_samples = len(data.images)
```

Splitting data on test and train

```python
from sklearn.model_selection import train_test_split
# Separating original data on test and train
X_train_og, X_test_og, y_train_og, y_test_og = train_test_split(data.images.reshape((n_samples, -1)), data['target'], test_size=0.3, shuffle=True)


# SVM
X_test_svm_og = X_test_og
y_test_svm_og = y_test_og

# KNN
X_test_knn_og = X_test_og
y_test_knn_og = y_test_og

# RF
X_test_rf_og = X_test_og
y_test_rf_og = y_test_og
```

#GA Gridsearch for SVM classifier#

```python
pip install sklearn-genetic-opt
```

```python
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn_genetic import GASearchCV
from sklearn.model_selection import train_test_split
from sklearn_genetic.space import Continuous, Categorical, Integer
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn_genetic.plots import plot_fitness_evolution
```

```python
parameters = {"kernel": Categorical(["linear", "rbf", "sigmoid", "poly"]), "C": Integer(1, 11), "class_weight": Categorical([None, "balanced"])}
svc = svm.SVC()
```

```python
clf = GASearchCV(estimator=svc, cv=2, param_grid=parameters, population_size=10, generations=40, n_jobs=4, verbose=3, criteria="max")
```

```python
clf.fit(X_train_og, y_train_og)
```

```python
clf.best_params_
```

```python
pred = clf.predict(X_test_svm_og)

print(precision_score(y_test_svm_og, pred, average="micro"))
```

```python
plot_fitness_evolution(clf, metric="fitness")
```

```python
plot_fitness_evolution(clf, metric="fitness_std")
```
